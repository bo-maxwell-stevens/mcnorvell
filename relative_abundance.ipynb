```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Relative abundance by plot

```python
# Load the datasets
df_18S = pd.read_csv('18S/relative_abundance.csv', index_col=0)
df_ITS = pd.read_csv('ITS/relative_abundance.csv', index_col=0)
```

```python
def extract_taxonomic_level(taxonomy, level):
    parts = taxonomy.split(';')
    if level < len(parts):
        part = parts[level].strip('_')
        return part if part else 'Unknown'
    return 'Unknown'
```

```python
# Extract genus for 18S (6th level: g__Genus)
df_18S['Genus'] = df_18S.index.map(lambda x: extract_taxonomic_level(x, 5))
grouped_18S = df_18S.groupby('Genus').sum()

# Normalize to percentages
grouped_18S = grouped_18S.div(grouped_18S.sum(axis=0), axis=1) * 100

# Extract family for ITS (5th level: f__Family)
df_ITS['Phylum'] = df_ITS.index.map(lambda x: extract_taxonomic_level(x, 1))
grouped_ITS = df_ITS.groupby('Phylum').sum()

# Normalize to percentages
grouped_ITS = grouped_ITS.div(grouped_ITS.sum(axis=0), axis=1) * 100
```

```python
# Plotting
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Stacked bar plot for 18S at genus level
grouped_18S.T.plot(kind='bar', stacked=True, ax=axes[0], legend=False)
axes[0].set_title('18S: Relative Abundance at Genus Level')
axes[0].set_ylabel('Relative Abundance (%)')
axes[0].set_xlabel('Samples')

# Stacked bar plot for ITS at family level
grouped_ITS.T.plot(kind='bar', stacked=True, ax=axes[1], legend=False)
axes[1].set_title('ITS: Relative Abundance at Family Level')
axes[1].set_ylabel('Relative Abundance (%)')
axes[1].set_xlabel('Samples')

# Adjusting legend and layout
axes[0].legend(title='Genus (18S)', bbox_to_anchor=(1.05, 1), loc='upper left')
axes[1].legend(title='Family (ITS)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.savefig('Figures/relative_abundance_by_plots.jpg', dpi = 300, bbox_inches = 'tight')
# Show plot
plt.show()
```

```python
# Transpose the grouped dataframes to align with mapping file
grouped_18S_transposed = grouped_18S.T
grouped_ITS_transposed = grouped_ITS.T

# Merge with mapping file
merged_18S = pd.merge(mapping_file, grouped_18S_transposed, left_on='sampleid', right_index=True)
merged_ITS = pd.merge(mapping_file, grouped_ITS_transposed, left_on='sampleid', right_index=True)

# Select numeric columns for grouping and normalization
numeric_columns_18S = merged_18S.select_dtypes(include='number')
numeric_columns_ITS = merged_ITS.select_dtypes(include='number')

# Group by 'fire' treatment and normalize
grouped_by_fire_18S = numeric_columns_18S.groupby(merged_18S['fire']).sum()
grouped_by_fire_18S = grouped_by_fire_18S.div(grouped_by_fire_18S.sum(axis=0), axis=1) * 100

grouped_by_fire_ITS = numeric_columns_ITS.groupby(merged_ITS['fire']).sum()
grouped_by_fire_ITS = grouped_by_fire_ITS.div(grouped_by_fire_ITS.sum(axis=0), axis=1) * 100

# Plotting in subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Stacked bar plot for 18S
grouped_by_fire_18S.plot(kind='bar', stacked=True, ax=axes[0], legend=False)
axes[0].set_title('18S: Relative Abundance by Fire Treatment')
axes[0].set_ylabel('Relative Abundance (%)')
axes[0].set_xlabel('Fire Treatment')

# Stacked bar plot for ITS
grouped_by_fire_ITS.plot(kind='bar', stacked=True, ax=axes[1], legend=False)
axes[1].set_title('ITS: Relative Abundance by Fire Treatment')
axes[1].set_ylabel('Relative Abundance (%)')
axes[1].set_xlabel('Fire Treatment')

# Adjusting legend and layout
axes[0].legend(title='Phylum (18S)', bbox_to_anchor=(1.05, 1), loc='upper left')
axes[1].legend(title='Phylum (ITS)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show plot
plt.show()
```
